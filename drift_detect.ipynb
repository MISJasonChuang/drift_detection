```python
import pandas as pd
import numpy as np
from skmultiflow.drift_detection.adwin import ADWIN
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Gold_Prices.csv')
df
```

```python
us = df[["US dollar"]]
print(us.isnull().sum())
us = us.to_numpy()
us
```

```python
adwin = ADWIN(delta=0.002)
# Adding stream elements to ADWIN and verifying if drift occurred
count = 0
for i in range(len(us)):
    adwin.add_element(us[i][0])
    if adwin.detected_change():
        count+=1
        print('Change detected in data: ' + str(us[i][0]) + ' - at index: ' + str(i))
print("count ",count)
```
