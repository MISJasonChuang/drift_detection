```python
import pandas as pd
import numpy as np
from skmultiflow.drift_detection.adwin import ADWIN
```

```python
def drift_detect(input: list, delta: float=0.002):
  adwin = ADWIN(delta=delta)
  count = 0
  for i in range(len(input)):
      # Adding stream elements to ADWIN and verifying if drift occurred
      adwin.add_element(input[i][0])
      if adwin.detected_change():
          count+=1
          print('Change detected in data: ' + str(input[i][0]) + ' - at index: ' + str(i))
  print('Change detected times: '+str(count))
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Gold_Prices.csv')
us = df[["US dollar"]]
print(us.isnull().sum())
us = us.to_numpy()
us
```

```python
d = 0.002
drift_detect(input=us, delta=d)
```
